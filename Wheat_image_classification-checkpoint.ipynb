```python
# Setting paths
data_path = 'c:\\Users\\USER\\wheat_images_classification\\data'
processed_path = 'processed_data'

# Creating processed data directories
for split in ['train', 'valid', 'test']:
    for class_name in ['aphid', 'fusarium_head_blight', 'healthy', 'mildew', 'yellow_rust']:
        os.makedirs(os.path.join(processed_path, split, class_name), exist_ok=True)
```

```python
#defining image processing function

def process_image(image_path, target_size=(224, 224)):
   
    # Load image
    img = Image.open(image_path)
    
    # Convert to RGB 
    if img.mode != 'RGB':
        img = img.convert('RGB')
    
    # Resize
    img = img.resize(target_size)
    
    return img
```

```python
#processing training set images

print("Processing training set:")
for class_name in ['aphid', 'fusarium_head_blight', 'healthy', 'mildew', 'yellow_rust']:
    # Set paths
    input_dir = os.path.join(data_path, 'train', class_name)
    output_dir = os.path.join(processed_path, 'train', class_name)
    
    # Process each image
    n_processed = 0
    for img_name in os.listdir(input_dir):
        if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Process and save
            img = process_image(os.path.join(input_dir, img_name))
            img.save(os.path.join(output_dir, f'processed_{img_name}'))
            n_processed += 1
    
    print(f"{class_name}: {n_processed} images")
```

```python
#processing validation set

print("\nProcessing validation set:")
for class_name in ['aphid', 'fusarium_head_blight', 'healthy', 'mildew', 'yellow_rust']:
    # Set paths
    input_dir = os.path.join(data_path, 'valid', class_name)
    output_dir = os.path.join(processed_path, 'valid', class_name)
    
    # Process each image
    n_processed = 0
    for img_name in os.listdir(input_dir):
        if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Process and save
            img = process_image(os.path.join(input_dir, img_name))
            img.save(os.path.join(output_dir, f'processed_{img_name}'))
            n_processed += 1
    
    print(f"{class_name}: {n_processed} images")
```

```python
#processing test set

print("\nProcessing test set:")
for class_name in ['aphid', 'fusarium_head_blight', 'healthy', 'mildew', 'yellow_rust']:
    # Set paths
    input_dir = os.path.join(data_path, 'test', class_name)
    output_dir = os.path.join(processed_path, 'test', class_name)
    
    # Process each image
    n_processed = 0
    for img_name in os.listdir(input_dir):
        if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
            # Process and save
            img = process_image(os.path.join(input_dir, img_name))
            img.save(os.path.join(output_dir, f'processed_{img_name}'))
            n_processed += 1
    
    print(f"{class_name}: {n_processed} images")
```

```python
def empty_balanced_dir():
    balanced_path = 'balanced_data'
    if os.path.exists(balanced_path):
        shutil.rmtree(balanced_path)
    print("Balanced directory emptied")

# %%
empty_balanced_dir()
```

```python

def copy_all_data(processed_path, balanced_path):
    # Create directories for all splits
    splits = ['train', 'valid', 'test']
    classes = ['aphid', 'fusarium_head_blight', 'healthy', 'mildew', 'yellow_rust']
    
    for split in splits:
        for class_name in classes:
            os.makedirs(os.path.join(balanced_path, split, class_name), exist_ok=True)
            
            # Copy files
            src_path = os.path.join(processed_path, split, class_name)
            dst_path = os.path.join(balanced_path, split, class_name)
            
            files = os.listdir(src_path)
            for file in files:
                shutil.copy2(
                    os.path.join(src_path, file),
                    os.path.join(dst_path, file)
                )
            print(f"{split}/{class_name}: {len(files)} images")

# Execute copying
processed_path = 'processed_data'
balanced_path = 'balanced_data'
copy_all_data(processed_path, balanced_path)
```

##### Model Development

     Model development with efficientNet NBO using Transfer Learning

```python
#load efficientNet with pretrained weights
def create_model():
    
    base_model = MobileNetV2(
        weights='imagenet',
        include_top=False,
        input_shape=(160, 160, 3)
    )
    
    #freeze all layers
    base_model.trainable = False
    
    x = GlobalAveragePooling2D()(base_model.output)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.3)(x)
    outputs = Dense(5, activation='softmax')(x)
    
    return Model(base_model.input, outputs)
```

```python
# create model
model = create_model()
model.compile(
    optimizer=Adam(learning_rate=0.0004),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

    Training setup

```python
#setup callbacks
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=6,          
        restore_best_weights=True
    ),
    ReduceLROnPlateau(
        monitor='val_accuracy',
        factor=0.5,         
        patience=3,
        min_lr=1e-6
    ),
    ModelCheckpoint(
        'best_model.keras',
        monitor='val_accuracy',
        save_best_only=True
    )
]
```

    Training model

```python
history = model.fit(
    train_generator,
    epochs=15,
    validation_data=valid_generator,
    callbacks=callbacks
)
```
